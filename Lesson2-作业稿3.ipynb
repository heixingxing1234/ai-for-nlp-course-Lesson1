```python
def search1(start, city_connection):   #connection_graph是字典
    list_start = []
    keys = list(city_connection.keys())
    for key in keys:
        for value in city_connection[key]:
            if start == value:
                list_start.append(key)
    return list_start        
```

```python
list_start1 = []
point_start1 = []
seen1 = set()
seen2 = set()
```

```python
def bfs(point,destination,city_connection):
    
    if search2(point) != 'intersection':
        
        if search1(point,city_connection) == search1(destination,city_connection):
            list_start1 + search1(point,city_connection)
            return list_start1    
        seen1.add(search1(point,city_connection))  
    
    if search2(point) == 'intersection':
        
        for lineA in search1(point,city_connection):             
            
            #ss = copy.deepcopy(search1(point,city_connection))
            
            #lineA = search1(point,city_connection).pop()
            
            #lineA1 = ss.pop()
        
            if lineA in seen1: continue
                
            if lineA == search1(destination,city_connection):
                list_start1.append(lineA)
                return list_start1
            else:
                list_start1.append(lineA)
                seen1.add(lineA)
       
    
    for lineA in search1(point,city_connection):
        #对每条线找交点，储存
        lineAa = city_connection[lineA]
        for point1 in lineAa:                                   # 找交点
            if point1 in seen2: continue
            if search2(point1) == 'intersection':
                point_start1.append(point1)
                seen2.add(point1) 
            
    #对每个交点，找线       
        
    new_point = point_start1.pop()
    bfs(new_point,destination,city_connection)       # 递归   
                    
    return list_start1


```
